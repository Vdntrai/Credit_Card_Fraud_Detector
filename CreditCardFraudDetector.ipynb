#### The Data is very Unbalanced so it will hinder in the Machine Learning Process

```python
legit=df[df.Class==0]
fraud=df[df.Class==1]
```

### UnderSampling to Deal with the unbalanced Dataset

```python
legit_sample=legit.sample(492)
```

```python
newdf=pd.concat([legit_sample,fraud],axis=0)
newdf
```

```python
X=newdf.drop(['Class'],axis=1)
Y=newdf['Class']
```

#### Train Test Split

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=69)
```

```python
model=LogisticRegression( max_iter=699)
```

```python
model.fit(X_train,Y_train)
```

```python
train_pred=model.predict(X_train)
train_acc=accuracy_score(train_pred,Y_train)
print(f'Train Accuracy:{train_acc*100:.2f}%')
```

```python
test_pred=model.predict(X_test)
test_acc=accuracy_score(test_pred,Y_test)
print(f'Test Accuracy:{test_acc*100:.2f}%')
```
